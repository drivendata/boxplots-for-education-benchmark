```python
data_root = "data/provided"
training_set = pd.read_csv(os.path.join(data_root, "TrainingData.csv"), 
                           index_col=0)

test_set_values = pd.read_csv(os.path.join(data_root, "TestData.csv"),
                              index_col=0, 
                              dtype=training_set.dtypes)

submission_format = pd.read_csv(os.path.join(data_root, "SubmissionFormat.csv"),
                                index_col=0)

# labels are the ones in the training set but not in the test set
labels = list(set(training_set.columns) - set(test_set_values.columns))
non_labels = list(test_set_values.columns)

# split training into predictors and labels
training_set_values = training_set[non_labels]
training_set_labels = training_set[labels]
```

## What, precisely, is our task?

Now that we know what the labels look like, we need to know what our actual goal is. Looking at the submission format file gives us an idea. We see that we actually want to predict a probability [0, 1] for each of the possible labels in each of the categories.

For example, in the `Pre_K` category, we need to assign a probability to each of the values (`PreK`, `Non PreK`, `NO_LABEL`) that this category can have.

Ultimately there are rows that ERS knows are hard to categorize. There is always some set of the data that they need to go back to the district for clarification on. If we can not only tell ERS which labels are most likely, but how confident we are in those labels, they can streamline the process of labeling these problem rows.

# Let's make this a machine learning problem!

We need to do some preprocessing in order to train models that look like the submission format. We'll do that quickly here. We basically need a set of binary columns rather than the categorical data that we currently have.

To do that, we will:

 - give each posssible label value in each column a distinct name
 - use `sklearn`'s `MultiLabelBinarizer` to create 0/1 columns

```python
from sklearn.preprocessing import MultiLabelBinarizer

# Creates long labels by prepending category so that we can
# treat all independently; for every column does the following:
#     Use                    Use
#     -----            ==>   -----
#     Instruction            Use__Instruction
#     NO_LABEL               Use__NO_LABEL
#
def expand_dataframe_labels(df, inplace=False, delimiter="__"):
    if not inplace:
        df = df.copy()
    
    for c in df.columns:
        def _prepend(x, label="NONE"):
            if pd.isnull(x):
                x = "NONE"
            return "{}{}{}".format(label, delimiter, str(x))

        df[c] = df[c].apply(_prepend, label=c)
        
    return df

# Wrap the above to go directly from our training set labels
# to a binary matrix that matches the submission format.
def multiple_label_columns_to_binary_matrix(data_frame):
    # create the expanded labels
    expanded_df = expand_dataframe_labels(data_frame)

    # turn in to binary matrix
    mlb = MultiLabelBinarizer()
    binary_label_matrix = mlb.fit_transform(expanded_df.values)
    
    return binary_label_matrix, mlb.classes_.tolist() 

# get the matrix and labels after processing
label_matrix, label_names = multiple_label_columns_to_binary_matrix(training_set_labels)

# create a dataframe for easy viewing
training_set_binary_labels = pd.DataFrame(label_matrix, 
                                          columns=label_names, 
                                          index=training_set_labels.index)

# double-check that our columns are in the same order as the submission format
assert((training_set_binary_labels.columns.values == submission_format.columns.values).all())

training_set_binary_labels.head()
```

# Great work on the labels, but what does the data look like?

Now's the time to take a look at what we'll be using to make predictions. There are 16 feature columns. These are all free-form text fields except for the `FTE` column (which is the percent of a 40hr work week the employee works) and the `Total` column (which is the dollar amount for this line item).

```python
print training_set_values.shape
training_set_values.head(10)
```

```python
# Concatenates all text fields into one big string.
#    to_lower: forces all text to be lowercase
#    to_drop: drops columns that aren't text fields
#
def flatten_to_text(data_frame, to_lower=True, to_drop=[]):
    # drop numeric columns
    text_data = data_frame.drop(to_drop, axis=1)
    
    # replace nans with blanks
    text_data[pd.isnull(text_data)] = ""
    
    # create a single string out of a row
    def _flatten_row(x):
        if to_lower:
            return " ".join(map(str.lower, x.values.flatten().tolist()))
        else:
            return " ".join(x.values.flatten().tolist())

    return text_data.apply(_flatten_row, axis=1)

# flatten text for training set
all_text = flatten_to_text(training_set_values, to_drop=["FTE", "Total"])

# flatten text for test set
all_text_test = flatten_to_text(test_set_values, to_drop=["FTE", "Total"])

# there are some identical rows, trim them for performance sake
all_text.drop_duplicates(inplace=True)
print all_text.shape

all_text.head()
```

# Features!

As with many NLP problems, the magic is in creating effective features. We'll start here with a simple strategy: [TF-IDF (term frequency, inverse document frequency)](http://en.wikipedia.org/wiki/Tf%E2%80%93idf).

`sklearn` has good tools to convert text to TF-IDF features, so we'll use that for our benchmarking.

[Naive bayes](http://en.wikipedia.org/wiki/Naive_Bayes_classifier) is a common first choice for a text classification problem, so we'll start with that.

```python
%%time
from sklearn.feature_extraction.text import TfidfVectorizer

vec = TfidfVectorizer(stop_words='english')

tfidf_features = vec.fit_transform(all_text)
```

```python
%%time
from sklearn.naive_bayes import BernoulliNB
from sklearn.multiclass import OneVsRestClassifier

nb_clf = OneVsRestClassifier(BernoulliNB())

result = nb_clf.fit(tfidf_features.toarray(), training_set_binary_labels.loc[all_text.index].values)
```

```python
# sklearn comes with joblib which efficiently pickles numpy arrays
# http://scikit-learn.org/stable/modules/model_persistence.html
from sklearn.externals import joblib
pickles = joblib.dump(nb_clf, 'pickles/naive_bayes.pkl')
```

```python
from sklearn.externals import joblib
nb_clf = joblib.load('pickles/naive_bayes.pkl')
```

# We've got our model, let's predict!

Now that we have trained our feature extraction and statistical model, we can predict on our test set by doing the following:
 - Use the `vec` object that we trained to `transform` our test set into our model features.
 - Call the `predict_proba` method on our trained clasifier `nb_clf` passing in the test set TF-IDF data

```python
# transform the test set into the tf-idf features
# note that we fit vec
tf_idf_features_test = vec.transform(all_text_test)
nb_preds = nb_clf.predict_proba(tf_idf_features_test)
```

```python
# takes a matrix of predictions and writes out
# a submission file in the proper format
def write_submission_file(predictions, filename):
    # we read submission_format from the csv file
    # already, so we will copy the headers from here
    submission_df = submission_format.copy()

    # transfer the probabilities we predicted into the
    # dataframe that has the right column names
    for i, c in enumerate(submission_df.columns):
        submission_df[c] = predictions[:,i].astype(np.float64)

    submission_df.to_csv(filename)
    
write_submission_file(nb_preds, "data/preds/NaiveBayes.csv")
```

# Let's get just a little bit smarter in our model

Naive Bayes was all good, but we're in a competition here. Time to break out the big guns.

First trick is to know that we're considering a specific scoring metric for the competition. While these are generally engineered to correlate with other ways of measuring success, tuning to this particular function may help our score.

Another note is that [logistic regression](http://en.wikipedia.org/wiki/Logistic_regression) produces well-calibrated probabilities. This means that our scores on a metric where probabilities matter should be better.

We can use logistic regression and a cross-validated grid search to improve our score. That is, we'll search over the space of parameters to a logistic regression model so that we can optimize directly the metric we care about. This parameter search is built in to `sklearn`, so it is cheap for us to use the work we've already done to produce a smarter model.

### Fist step is to define the metric that we want to optimize
The code for this metric is provided to competitors [on our github repo](https://github.com/drivendataorg/metrics).

```python
# Defined for your convenience; these are the
# class_column_indices for the Box-Plots for Education competition
# www.drivendata.org/competitions/4/
#
# CODE FROM: https://github.com/drivendataorg/metrics
#
BOX_PLOTS_COLUMN_INDICES = [range(37),
                            range(37,48),
                            range(48,51),
                            range(51,76),
                            range(76,79),
                            range(79,82),
                            range(82,87),
                            range(87,96),
                            range(96,104)]

def multi_multi_log_loss(predicted, actual, class_column_indices, eps=1e-15):
    """Multi class, multi-label version of Logarithmic Loss metric.
    :param predicted: a 2d numpy array of the predictions that are probabilities [0, 1]
    :param actual: a 2d numpy array of the same shape as your predictions. 1 for the actual labels, 0 elsewhere 
    :return: The multi-multi log loss score for this set of predictions
    """
    class_scores = np.ones(len(class_column_indices), dtype=np.float64)

    # calculate log loss for each set of columns that belong to a class:
    for k, this_class_indices in enumerate(class_column_indices):
        # get just the columns for this class
        preds_k = predicted[:, this_class_indices]

        # normalize so probabilities sum to one (unless sum is zero, then we clip)
        preds_k /= np.clip(preds_k.sum(axis=1).reshape(-1, 1), eps, np.inf)

        actual_k = actual[:, this_class_indices]

        # shrink predictions
        y_hats = np.clip(preds_k, eps, 1 - eps)
        sum_logs = np.sum(actual_k * np.log(y_hats))
        class_scores[k] = (-1.0 / actual.shape[0]) * sum_logs

    return np.average(class_scores)
```

### Convert the metric code above to an `sklearn` compatible function

scikit-learn let's you definte a custom function to optimize. Using the [documentation for their `make_scorer` method](http://scikit-learn.org/stable/modules/generated/sklearn.metrics.make_scorer.html), we can wrap the provided `multi_mult_log_loss` method in a way that `sklearn` can call while it is doing cross-validation.

```python
# adjust api to make a custom scorer for sklearn
from sklearn.metrics import make_scorer

# Note: don't make this a lambda, since that won't pickle
def sklearn_multi_multi(actual, predicted):
    return multi_multi_log_loss(predicted, actual, BOX_PLOTS_COLUMN_INDICES)

multi_log_loss_scorer = make_scorer(sklearn_multi_multi, greater_is_better=False, needs_proba=True)
```

### Train our logistic regression model over a space of possible parameters

```python
%%time
# sklearn issue causes some annoying warnings, so we turn them off
import warnings
warnings.filterwarnings('ignore')

from sklearn.linear_model import LogisticRegression
from sklearn.grid_search import GridSearchCV

lr_clf = OneVsRestClassifier(LogisticRegression())
    
lr_grid = GridSearchCV(lr_clf,
                       # note "estimator__" is so these passed through OneVsRestClassifier
                       # down to LogisticRegression
                       param_grid={'estimator__penalty': ['l1', 'l2'], 
                                   'estimator__C': [0.1, 1.0],
                                   'estimator__fit_intercept': [True, False]}, 
                       scoring=multi_log_loss_scorer)

lr_grid.fit(tfidf_features.toarray(), training_set_binary_labels.loc[all_text.index].values)
```

```python
pickles = joblib.dump(lr_grid, 'pickles/logistic_reg_grid.pkl')
```

```python
lr_grid = joblib.load('pickles/logistic_reg_grid.pkl')
```

```python
lr_preds = lr_grid.predict_proba(tf_idf_features_test)
write_submission_file(lr_preds, "data/preds/LogisticRegression.csv")
```
